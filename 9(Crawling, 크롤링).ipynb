```python
import requests
from bs4 import BeautifulSoup
```

```python
# Crawling 기본 모델
url = 'https://www.imdb.com/chart/top/?ref_=hm_nv_menu'
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

titles = soup.find_all('h3', class_='ipc-title__text ipc-title__text--reduced')
for t in titles:
    print(t.get_text())
    
# Crawling에서 필수적인 작업이 웹 페이지에서 HTML 코드 열고 가져오고 싶은 부분(제목/본문 등) 'tag' 'class' 확인해서 내 코드에 대입하는 것.
```

```python
url = 'https://www.skku.edu/skku/campus/skk_comm/notice01.do'
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

titles = soup.find_all('a', attrs={'title': '자세히 보기'}) # 제목에 하이퍼링크 걸려있는 것들은 class가 없음. 'attrs~'로 보완.
for i in titles:
    print(i.get_text())
```
